```python
# =====================================================================
# ||                                                                 ||
# ||                           Backtesting                           ||
# ||                                                                 ||
#======================================================================

df['Change'] = df['Close'].pct_change() * 100

STARTING_BALANCE = 1000
MINIMUM_PRICE = 100
ETORO_REGULAR_FEE = 0.00012

position_counter = 0

wallet = Wallet(STARTING_BALANCE)

benchmark_performances = np.zeros(len(df))
first_buy = None

for i, date in enumerate(df.index):
    buy = False
    position = 0
    if df.loc[date]['SEFI Signal'] or df.loc[date]['ADR_signal']:
        position_price = df.loc[date]['Close']
        position_price_plus_fee = position_price * ETORO_REGULAR_FEE

        if wallet.balance > 100:
            position = Position(100, 1, sp500_database)
            wallet.add_position(position)
            buy = True
            if not first_buy:
                first_buy = date
            position_counter += 1

    if not buy:
        if i > 0:
            benchmark_performances[i] = df['Change'].loc[date] * position_counter
        else:
            benchmark_performances[i] = 0
    else:
        if not first_buy:
            benchmark_performances[i] = 0
        else:
            benchmark_performances[i] = df['Change'].loc[date] * position_counter
```

```python
# (Ending price - starting price)/(Starting price)
position_counter
```

```python
benchmark_performances = benchmark_performances.cumsum()
print(benchmark_performances)
```

```python
import matplotlib.pyplot as plt

benchmark_start = df.index.get_loc(first_buy)

df['benchmark'] = [0 if i < benchmark_start else df.iloc[i]['Change'] for i, date in enumerate(df.index)]
df['benchmark'] = df['benchmark'].cumsum()

plt.figure(figsize=(12, 6), dpi=200)

plt.plot(df.index, df['Close'], label="PRICE")
for date in df[(df['SEFI Signal'] == True) | (df['ADR_signal'] == True)].index:
    plt.plot(date, df['Close'].loc[date], marker="o", c='g')
```

```python
plt.figure(figsize=(12, 6), dpi=200)

plt.plot(range(len(benchmark_performances)), benchmark_performances, label="PERFORMANCE")
plt.plot(range(len(df)), df['benchmark'].fillna(0), label="BENCHMARK")
plt.legend()
```
