```python
import yfinance as yf
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
df = yf.download("^GSPC", period="2y", interval="1d")
df['Change'] = (df['Close'].pct_change() * 100)
df["Change MA20"] = df['Change'].rolling(window=20).mean()
```

```python
df.dropna(inplace=True)
results = seasonal_decompose(df['Change'], period=5)
results.trend.plot(figsize=(12, 6), grid=True)
```

```python
results.seasonal.plot(figsize=(16, 6), xlim=["2021-01", "2021-12"])
```
