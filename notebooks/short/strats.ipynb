```python
from classes.market_analysis.SP500Analysis import SP500Analysis
from classes.database.SP500Database import SP500Database
from config import db_path
import datetime
```

```python
sp500_database = SP500Database()
sp500_database.connect_existing_database(db_path / "sp500.sqlite")
market_analysis = SP500Analysis(sp500_database)
sefi = market_analysis.sefi()
adr = market_analysis.adr_analysis()
dates = sp500_database.query_all_dates()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6), dpi=200)
plt.plot(market_analysis.sp500.index, market_analysis.sp500['Close'])
for date in market_analysis.sp500[
    (market_analysis.sp500['ADR Signal Short']) | (market_analysis.sp500['SEFI Signal Short'])].index:
    plt.plot(date, market_analysis.sp500.loc[date]['Close'], marker='o', c='r')
for date in market_analysis.sp500[
    (market_analysis.sp500['ADR Signal Long']) | (market_analysis.sp500['SEFI Signal Long'])].index:
    plt.plot(date, market_analysis.sp500.loc[date]['Close'], marker='o', c='g')
```
