```python
ids = []
for i in range(len(df)):
    ids.append(i)
```

```python
df["Id"] = ids
```

```python
df.drop("Id", axis=1, inplace=True)
```

```python
df
```

```python
df['Change'] = df['Adj Close'].pct_change() * 100
```

```python
df.dropna(inplace=True)
```

```python
df
```

```python
def get_signal(change):
    if change < 0:
        return True
    return False
```

```python
df['Signal'] = np.vectorize(get_signal)(df['Change'])
```

```python
df
```

```python
df[(df['Signal'] == False) & (df['Change'] > 1.5)]
```
